# Metrics to Impute missing values

Analysis for imputing

1. AMT_REQ_CREDIT_BUREAU (HOUR, DAY, WEEK, MON, QRT, YEAR) = 13.50%

```python
print(application_data['AMT_REQ_CREDIT_BUREAU_HOUR'].value_counts())
print(application_data['AMT_REQ_CREDIT_BUREAU_DAY'].value_counts())
print(application_data['AMT_REQ_CREDIT_BUREAU_WEEK'].value_counts())
print(application_data['AMT_REQ_CREDIT_BUREAU_MON'].value_counts())
print(application_data['AMT_REQ_CREDIT_BUREAU_QRT'].value_counts())
print(application_data['AMT_REQ_CREDIT_BUREAU_YEAR'].value_counts())
```

Considering the data in the columns, we can assume that data to be categorical. We can categorise loan application where AMT_REQ_CREDIT_BUREAU_HOUR is [0-4],AMT_REQ_CREDIT_BUREAU_DAY is [0-9], AMT_REQ_CREDIT_BUREAU_WEEK is [0-8], AMT_REQ_CREDIT_BUREAU_MON is [0-27], AMT_REQ_CREDIT_BUREAU_QRT is [0-261] and AMT_REQ_CREDIT_BUREAU_YEAR is [0-25].

# If we see categorywise, maximum(more than 85%) applicants have '0'. Hence , we can impute null values with mode

```python
replace_AMT_REQ_CREDIT_BUREAU_HOUR = application_data['AMT_REQ_CREDIT_BUREAU_HOUR'].mode()[0]
print('AMT_REQ_CREDIT_BUREAU_HOUR',replace_AMT_REQ_CREDIT_BUREAU_HOUR)
replace_AMT_REQ_CREDIT_BUREAU_DAY = application_data['AMT_REQ_CREDIT_BUREAU_DAY'].mode()[0]
print('AMT_REQ_CREDIT_BUREAU_DAY',replace_AMT_REQ_CREDIT_BUREAU_DAY)
replace_AMT_REQ_CREDIT_BUREAU_WEEK = application_data['AMT_REQ_CREDIT_BUREAU_WEEK'].mode()[0]
print('AMT_REQ_CREDIT_BUREAU_WEEK',replace_AMT_REQ_CREDIT_BUREAU_WEEK)
replace_AMT_REQ_CREDIT_BUREAU_MON = application_data['AMT_REQ_CREDIT_BUREAU_MON'].mode()[0]
print('AMT_REQ_CREDIT_BUREAU_MON',replace_AMT_REQ_CREDIT_BUREAU_MON)
replace_AMT_REQ_CREDIT_BUREAU_QRT = application_data['AMT_REQ_CREDIT_BUREAU_QRT'].mode()[0]
print('AMT_REQ_CREDIT_BUREAU_QRT',replace_AMT_REQ_CREDIT_BUREAU_QRT)
replace_AMT_REQ_CREDIT_BUREAU_YEAR = application_data['AMT_REQ_CREDIT_BUREAU_YEAR'].mode()[0]
print('AMT_REQ_CREDIT_BUREAU_YEAR',replace_AMT_REQ_CREDIT_BUREAU_YEAR)
```

# Filling the missing values with mode values calculated i.e. 0.0

```python
application_data['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(replace_AMT_REQ_CREDIT_BUREAU_HOUR,inplace = True)
application_data['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(replace_AMT_REQ_CREDIT_BUREAU_DAY,inplace = True)
application_data['AMT_REQ_CREDIT_BUREAU_WEEK'].fillna(replace_AMT_REQ_CREDIT_BUREAU_WEEK,inplace = True)
application_data['AMT_REQ_CREDIT_BUREAU_MON'].fillna(replace_AMT_REQ_CREDIT_BUREAU_MON,inplace = True)
application_data['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(replace_AMT_REQ_CREDIT_BUREAU_QRT,inplace = True)
application_data['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(replace_AMT_REQ_CREDIT_BUREAU_YEAR,inplace = True)
```

```python
def age(x) :
    if x <= 30 :
        return '20s'
    if x<= 40 :
        return '30s'
    if x<= 50 :
        return '40s'
    if x<= 60 :
        return '50s'
    if x>= 61 :
        return 'Senior_Citizen'
application_data['Bin_DAYS_BIRTH'] = application_data['DAYS_BIRTH'].apply(age)
```

2. AMT_CREDIT

```python
def category(x) : 
    if x <= 250000 :
        return 'Low'
    if (x > 250000) & (x <= 500000) :
        return 'Average'
    if (x > 500000) & (x <= 800000) :
        return 'Good'
    if x > 800000 :
        return 'Best'
application_data['Bin_AMT_CREDIT'] = application_data['AMT_CREDIT'].apply(category)
application_data.head(5)
```

```python
plt.title('Non Defaulter')
application_data[application_data['TARGET']== 0].Bin_AMT_CREDIT.value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.title('Defaulter')
application_data[application_data['TARGET']== 1].Bin_AMT_CREDIT.value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

1.There is no significant difference in the distribution of% of AMT_CREDIT to the applicant with payment problems or not.
2. However, the highest number of applicants with payment problems is below the Central category ((Amt_credit> 500000) & (Amt_credit <= 800000)
